```python
from dataclasses import dataclass, field

@dataclass
class QuantitativeHistogram:
    
    project: str
    experiment: str
    compound: str = field(default=None)
    region: str = field(default=None)
    compounds: list = field(default=None)
    regions: list = field(default=None)
    
    def __post_init__(self):
        print(self.project, self.experiment)
        if self.compounds and isinstance(self.compounds, list) and self.region and isinstance(self.region, str):
            print(f"plotting compounds {self.compounds} in region {self.region}")
        elif self.compound  and isinstance(self.region, str) and self.regions and  isinstance(self.compounds, list):
            print(f"plotting compound {self.compound} in regions {self.regions} in ")
        else:
            raise ValueError("Please specify at least one compound and one region")
        
QuantitativeHistogram('TEST', 'TEST', regions=['OF'], compounds='5HTP')
```
